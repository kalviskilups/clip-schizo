```python
import defdevice
defdevice.force_device("cuda:3")
import torch
import torch.nn.functional as f
import math
import numpy as np

max_pool2d = nn.MaxPool2d(2, stride=2, return_indices=True)
max_unpool2d = nn.MaxUnpool2d(2, stride=2)

class Model():
    def __init__(s):
        s.parameters = []
    
    def bias(s, outs):
        tens = torch.zeros([outs], dtype = torch.float32, requires_grad = True, device = defdevice.def_device)
        s.parameters.append(tens)
        return tens
    
    def weight(s, ins, outs, variance_rescaling = 0.97, activation_sd_rescaling = 1.0):
        k = math.sqrt(3 * variance_rescaling / ins) / activation_sd_rescaling
        data = np.random.uniform(-k, k, (ins, outs))
        tens = torch.tensor(data, dtype = torch.float32, requires_grad = True, device =defdevice.def_device)
        s.parameters.append(tens)
        return tens
        
    def conv(s, size, ins, outs, variance_rescaling = 0.97, activation_sd_rescaling = 1.0):
        count = size * size * ins
        k = math.sqrt(3 * variance_rescaling / count) / activation_sd_rescaling
        data = np.random.uniform(-k, k, (outs, ins, size, size))
        tens = torch.tensor(data, dtype = torch.float32, requires_grad = True, device = defdevice.def_device)
        s.parameters.append(tens)
        return tens
    
    def deconv(s, size, ins, outs, variance_rescaling = 0.97, activation_sd_rescaling = 1.0):
        count = size * size * ins
        k = math.sqrt(3 * variance_rescaling / count) / activation_sd_rescaling
        data = np.random.uniform(-k, k, (ins, outs, size, size))
        tens = torch.tensor(data, dtype = torch.float32, requires_grad = True, device = defdevice.def_device)
        s.parameters.append(tens)
        return tens
    
    def forward(s, x):
        raise NotImplemented()
```

```python
class UltraAutoencoder(Model):
    def __init__(s, channels = [3, 32]):
        super().__init__()        
        
        s.autoenc_levels = len(channels) - 1
        s.channels = channels
        
        s.convs = []
        s.tconvs = []
        s.b = []
        s.tb = []
        s.deconv_evid = []
        s.deconv_val = []
        s.generic = []
        for i in range(s.autoenc_levels):
            s.convs.append(s.conv(5, s.channels[i], s.channels[i + 1], activation_sd_rescaling = 1.5))
            s.tconvs.append(s.conv(1, s.channels[i + 1], s.channels[i + 1], activation_sd_rescaling = 1.5))
            s.b.append(s.bias(s.channels[i + 1]))
            s.tb.append(s.bias(s.channels[i + 1]))
            
            s.generic.append(s.bias(s.channels[i + 1]))
            
            s.deconv_evid.append(s.deconv(3, s.channels[i + 1], s.channels[i]))
            s.deconv_val.append(s.deconv(3, s.channels[i + 1], s.channels[i]))
        
        s.paramsets = [[s.convs[i], s.tconvs[i], s.generic[i], s.b[i], s.tb[i], s.deconv_evid[i], s.deconv_val[i]] for i in range(len(s.convs))]
        
        
    def forward(s, x, dropout = False, noise = True):
        orig = x
        ox = orig.shape[2]
        oy = orig.shape[3]
        pad_grid = 2 ** s.autoenc_levels
        x = f.pad(x, (0, (pad_grid - x.shape[3] % pad_grid) % pad_grid, 0, (pad_grid - x.shape[2] % pad_grid) % pad_grid))
        s.o = [x]
        s.ph = []
        s.h = []
        s.ind = []
        s.dh = []
        s.exp = []
        s.val = []
        s.norm = []
        s.d = []
        s.losses = []
        for i in range(s.autoenc_levels):
            if noise:
                s.ph.append(f.conv2d(s.o[-1] + torch.randn_like(s.o[-1]) / 8, s.convs[i], padding = "same"))
            else:
                s.ph.append(f.conv2d(s.o[-1], s.convs[i], padding = "same"))
            if dropout:
                s.ph[-1] = f.dropout(s.ph[-1], 0.3) / (1 / (1 - 0.3))
            s.ph[-1] = s.ph[-1] + torch.tanh(s.b[i]).reshape([1, -1, 1, 1])
            s.ph[-1] = torch.sigmoid(s.ph[-1])
            s.ph[-1] = (f.conv2d(s.ph[-1], s.tconvs[i], padding = 'same') + torch.tanh(s.tb[i]).reshape([1, -1, 1, 1]))
            s.ph[-1] = torch.sigmoid(s.ph[-1])
            
            #small_loss = torch.relu(0.4 - (s.ph[-1] ** 2).mean(dim = 3).mean(dim = 2).mean()) ** 2
            #big_loss = (torch.relu(s.ph[-1] - 0.99) ** 2).mean() * 1000
            #mini_loss = (torch.relu(- s.ph[-1] + 0.01) ** 2).mean() * 1000
            #s.curses = [small_loss, big_loss, mini_loss]
            
            h, ind = max_pool2d(s.ph[-1])
            s.h.append(h)
            s.ind.append(ind)
            s.o.append(s.h[-1])
            
            s.dh.append(max_unpool2d(s.h[-1], s.ind[-1]))
            s.dh[-1] = s.dh[-1] * ((torch.abs(s.generic[i] + 0.00001) * 0.001 + 0.00001).clamp(-0.1, 1).reshape([1, -1, 1, 1]))
            
            s.d.append(torch.sigmoid(f.conv_transpose2d(s.dh[-1], s.deconv_val[i])[:, :, 1:-1, 1:-1]))
            
            #s.exp.append(torch.exp(torch.sigmoid(s.deconv_evid[i] / 3)))
            #s.val.append(f.conv_transpose2d(s.dh[-1], s.deconv_val[i] * s.exp[-1]))
            
            #s.norm.append(f.conv_transpose2d(s.dh[-1], s.exp[-1]))
            #s.d.append((s.val[-1] / (s.norm[-1] + 0.001))[:, :, 1:-1, 1:-1])
            
            s.losses.append(((s.o[-2][:, :, :, :] - s.d[-1][:, :, :, :]) ** 2).mean(dim = 3).mean(dim = 2).mean())
        
        return s.d[0]
```

```python
class BetterModel(Model):
    def __init__(s, channels = [3, 4, 8]):
        super().__init__()
        
        s.channels = channels
        s.hidden_count = len(s.channels) - 1
        
        s.convs = []
        s.b = []
        for layer in range(s.hidden_count):
            s.convs.append(s.conv(4, s.channels[layer], s.channels[layer + 1]))
            s.b.append(s.bias(s.channels[layer + 1]))
        
        s.deconvs = []
        s.db = []
        s.sw = []
        s.sb = []
        s.sw2 = []
        s.sb2 = []
        for layer in range(s.hidden_count):
            s.deconvs.append(s.deconv(4, s.channels[layer + 1], s.channels[layer + 1]))
            s.db.append(s.bias(s.channels[layer + 1]))
            s.sw.append(s.conv(1, s.channels[layer + 1] + s.channels[layer], s.channels[layer], variance_rescaling = 4))
            s.sb.append(s.bias(s.channels[layer]))
            s.sw2.append(s.conv(1, s.channels[layer], s.channels[layer], variance_rescaling = 4))
            s.sb2.append(s.bias(s.channels[layer]))
            
            
        s.paramsets = []
        for layer in range(s.hidden_count):
            s.paramsets.append([s.convs[layer], s.b[layer], s.deconvs[layer], s.db[layer], s.sw[layer], s.sb[layer]])
    
    def forward(s, x):
        pad_grid = 4 ** s.hidden_count
        x = f.pad(x, (0, (pad_grid - x.shape[3] % pad_grid) % pad_grid, 0, (pad_grid - x.shape[2] % pad_grid) % pad_grid))
        s.inputs = [x]
        s.outputs = []
        for layer in range(s.hidden_count):
            conv = f.conv2d(s.inputs[-1], s.convs[layer], s.b[layer], stride = 4)
            s.inputs.append(torch.sigmoid(conv))
            
            deconv = f.conv_transpose2d(s.inputs[-1], s.deconvs[layer], s.db[layer], stride = 4)
            c = torch.cat((f.dropout(s.inputs[-2], p = 0.5) / 2, deconv), axis = 1)
            c = torch.sigmoid(f.conv2d(c, s.sw[layer], s.sb[layer], stride = 1))
            c = f.conv2d(c, s.sw2[layer], s.sb2[layer], stride = 1)
            s.outputs.append(c)
            
        s.losses = []
        for layer in range(s.hidden_count):
            s.losses.append(((s.inputs[layer] - s.outputs[layer]) ** 2).mean())
```

```python
class RandomModel(Model):
    def __init__(s, channels = [3, 8]):
        super().__init__()
        
        s.channels = channels
        s.hidden_count = len(s.channels) - 1
        
        s.convs = []
        s.b = []
        
        for layer in range(s.hidden_count):
            s.convs.append(s.conv(4, s.channels[layer], s.channels[layer + 1]))
            s.b.append(s.bias(s.channels[layer + 1]))
        
        s.deconvs = []
        for layer in range(s.hidden_count):
            s.deconvs.append(s.deconv(4, s.channels[layer + 1], s.channels[layer]))

        s.paramsets = []
        for layer in range(s.hidden_count):
            s.paramsets.append([s.convs[layer], s.b[layer], s.deconvs[layer]])

    def forward(s, x):
        pad_grid = 4 ** s.hidden_count
        x = f.pad(x, (0, (pad_grid - x.shape[3] % pad_grid) % pad_grid, 0, (pad_grid - x.shape[2] % pad_grid) % pad_grid))
        s.inputs = [x]
        s.outputs = []

        for layer in range(s.hidden_count):
            conv = f.conv2d(s.inputs[-1], s.convs[layer], s.b[layer], stride = 4)
            s.inputs.append(torch.softmax(conv, dim = 1))

            deconv = f.conv_transpose2d(s.inputs[-1], s.deconvs[layer], stride = 4)
            s.outputs.append(deconv)

        s.losses = []
        for layer in range(s.hidden_count):
            s.losses.append(((s.inputs[layer] - s.outputs[layer]) ** 2).mean())            
```

```python
def get_embs(image, model, size = 12):
    _ = model.forward(image.to('cuda:3'))
    res = []
    for h in model.inputs[1:]:
        res.append(tnp(f.interpolate(h, size = (size, size), mode = 'bilinear')[0]))
    res = np.concatenate(res, axis = 0)
    return res
```

```python
embs = []
for i in range(10000):
    embs.append(get_embs(imgs[i], model))
```

```python
image_ds = embdataset.RAMLoadDataset('COCOM2F_ALL_d8', load_amount = 10000)
text_ds = embdataset.RAMLoadDataset('COCOCLIP_ALL', load_amount = 10000)
```

```python
image_ds.sample_count = 10000
text_ds.sample_count = 10000
```

```python
image_ds.data = torch.tensor(np.stack(embs, axis = 0), dtype= torch.float32, requires_grad = False, device = 'cpu')
```

```python
def_device = 'cuda:3'

model2 = nns.Linear(512, sum(model.channels[1:]))

image_ds.reset()
text_ds.reset()

import torch
import torch.nn.functional

batch_count = 20000
batch_size = 128
optim = torch.optim.Adam(model2.parameters, lr = 0.0003)

image_noise = 0.07 * 3
text_noise = 0.15

for batch_idx in range(batch_count):
    ibatch = image_ds.get_batch(batch_size).to(def_device)
    tbatch = text_ds.get_batch(batch_size).to(def_device)
    
    ibatch = ibatch + torch.rand_like(ibatch) * image_noise
    tbatch = tbatch + torch.rand_like(tbatch) * text_noise
    
    tbatch = model2.forward(tbatch)
    
    neg_ibatch, neg_tbatch = losses.sample_negatives(ibatch, tbatch, batch_size)
    
    prods = losses.get_products(ibatch, tbatch)
    probs = torch.sigmoid(prods)
    
    neg_prods = losses.get_products(neg_ibatch, neg_tbatch)
    neg_probs = torch.sigmoid(neg_prods)
    
    loss_positive = losses.max_pressure_loss(probs, 1.0)
    loss_negative = losses.max_pressure_loss(neg_probs, 0.0)
    
    loss = (loss_positive + loss_negative) / 2
    
    loss.backward()
    optim.step()
    optim.zero_grad()
    
    if batch_idx % 1000 == 0:
        print(batch_idx, loss_positive, loss_negative)
```

```python
colorsl = [[255, 0, 0], [0, 255, 0], [0, 0, 255], [255, 255, 0], [255, 0, 255], [0, 255, 255], [0, 0, 0], [255, 255, 255], [128, 128, 128], [128, 0, 0], [0, 128, 0], [0, 0, 128]]
colors = np.array(colorsl)
colorsl = [[c / 255 for c in color] for color in colorsl]

def array_to_rgb(array):
    n, y, x = array.shape
    max_indices = np.argmax(array, axis=0)
    rgb_image = np.zeros((y, x, 3), dtype=np.uint8)
    for i in range(n):
        mask = max_indices == i
        rgb_image[mask] = colors[i]
        
    return rgb_image
```

```python
text_embr = embedders.CLIPTextEmbedder()

def run_image(path, labels):
    path = path
    img = imageops.open_image(path)
    hmaps = torch.tensor(get_embs(torch.tensor(img, device = defdevice.def_device).permute([2, 0, 1]).unsqueeze(0) / 255, model, 256), device = defdevice.def_device)
    lembs = text_embr.forward(labels)
    lembs = model2.forward(torch.tensor(lembs, device = 'cuda:3'))
    print(hmaps.shape)
    print(lembs.shape)
    prods = torch.einsum('eyx,ne->nyx', hmaps, lembs)
    hmaps = tnp(torch.softmax(prods, dim = 0))

    fig = plt.figure(figsize=(28, 8))
    
    width = ((len(labels) + 2 + 1) // 2) * 2
    gs = gridspec.GridSpec(2, width, width_ratios=[1, 0.05]*(width // 2))
    
    # Original image
    ax0 = fig.add_subplot(gs[0, 0])
    ax0.imshow(img, aspect = 'auto')
    ax0.set_xticks([])
    ax0.set_yticks([])

    ax1 = fig.add_subplot(gs[0, 2])
    ax1.imshow(array_to_rgb(hmaps), aspect = 'auto')
    ax1.set_xticks([])
    ax1.set_yticks([])
    
    legend_patches = [mpatches.Patch(color=color, label=label) for color, label in zip(colorsl[:len(labels)], labels)]
    ax1.legend(handles=legend_patches)

    # Individual heatmaps and their colorbars
    for idx, l in enumerate(labels):
        ax = fig.add_subplot(gs[(idx * 2 + 4) // width, (idx*2 + 4) % width])
        cax = fig.add_subplot(gs[(idx * 2 + 5) // width, (idx*2 + 5) % width])
        pos = cax.get_position()
        new_pos = [pos.x0 - 0.01, pos.y0, pos.width, pos.height]
        cax.set_position(new_pos)
        ax.set_xticks([])
        ax.set_yticks([])
        
        im = ax.imshow(hmaps[idx], aspect = 'auto', vmin = 0, vmax = 1)
        plt.colorbar(im, cax=cax, pad = 0)
        ax.set_title(l)

    plt.savefig(path[:-4] + 'output' + '.png', bbox_inches = 'tight')
```

```python
file_list = ['./RoboPhoto/1.jpg', './RoboPhoto/2.jpg', './RoboPhoto/3.jpg', './RoboPhoto/4.jpg'] #Images to run model on.
labels = ['roomba / metal box / warehouse robot', 'warehouse shelf', 'cardboard box', 'floor', 'ceiling', 'other'] #Labels to look for.

for path in file_list:
    run_image(path, labels)
```
